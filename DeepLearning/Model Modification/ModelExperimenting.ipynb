```python
#Previous Results not satisfactory enough
#low amount of change via epochs and high loss.
#model perhaps overfit. First new change did not work.

#importing required dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf
```

```python
#loading old data
usable_df = pd.read_csv("../FBI_csv_object_encoded.csv")
usable_df.head()
```

```python
#removing the index
usable_df = usable_df.drop(columns="Unnamed: 0",axis= 1)
usable_df.head()
```

```python
#re-breaking up the data.
#SPLIT THE DATA FOR TESTING AND TRAINING
y = usable_df["recovered_flag"].values
X = usable_df.drop(labels ="recovered_flag",axis =1).values
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=78)
```

```python
# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model,previous models may have been overfit, so will be toning down the number of neurons
# 3-3-1
#since the epochs stagnated early, will start with a smaller nubmer and see how it goes
number_input_features = len(X_train[0])
hidden_nodes_layer1 = 3
hidden_nodes_layer2 = 3

nn3 = tf.keras.models.Sequential()

# First hidden layer
nn3.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features,activation="relu"))
       
# Second hidden layer
nn3.add(tf.keras.layers.Dense(units=hidden_nodes_layer2 ,activation="relu"))

# Output layer
nn3.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn3.summary()
```

```python
# Compile the model
nn3.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
#send it
fit_model = nn3.fit(X_train,y_train,epochs=50)
```

```python
# Evaluate the model using the test data
model_loss3, model_accuracy3 = nn3.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss3}, Accuracy: {model_accuracy3}")
#look at that, a slight increase in accuracy and .10 down in loss!
#maybe in the right direction
#reduced number of neurons/layers and epochs
#still noticing stagnation during training.
```

```python
#export current model to HDF5 File
nn3.save("FBI_Model_Save3.h5")
```
