### Read the original CSV and Perform Basic Data Cleaning

```python
crimeD_df = crime_df.drop(columns=['ori','pub_agency_unit','data_year.1','prop_desc_code.1', 'date_recovered'])
print(crimeD_df.shape)
crimeD_df.head()    

# ORIGINATING AGENCY IDENTIFIER (ORI)
```

```python
# Replace empty value with logical value
crimeD_df['offender_race'].fillna('Unknown', inplace = True)
crimeD_df['offender_ethnicity'].fillna('Unknown', inplace = True)
crimeD_df['offender_age'].fillna(0.0, inplace = True)
crimeD_df['offender_sex'].fillna('U', inplace = True)
crimeD_df['weapon_name'].fillna('Unknown', inplace = True)
crimeD_df['weapon_code'].fillna(95, inplace = True)
crimeD_df['stolen_value'].fillna(0, inplace = True)
```

```python
## OW - Take a closer look at the entries with undefined population_group_description values.
crimeD_df.loc[crimeD_df.population_group_desc.isna() == True]
```

```python
## OW 
crimeD_df.population_group_desc = crimeD_df.population_group_desc.fillna("National")
crimeD_df.population_group_code = crimeD_df.population_group_code.fillna("0")
```

```python
crimeD_df.isna().sum()
```

```python
# What are the offenses?
crimeD_df.offense_name.value_counts()
```

```python
# OW - The priority for this project is anything related to stealing something.
excluded = ["Destruction/Damage/Vandalism of Property",
            "Drug/Narcotic Violations",
            "Impersonation",
            "Weapon Law Violations",
            "Drug Equipment Violations",
            "Simple Assault",
            "Aggravated Assault",
            "Kidnapping/Abduction",
            "Intimidation",
            "Extortion/Blackmail",
            "Rape",
            "Murder and Nonnegligent Manslaughter",
            "Arson",
            "Hacking/Computer Invasion",
            "Fondling",
            "Pornography/Obscene Material",
            "Human Trafficking, Involuntary Servitude",
            "Sodomy",
            "Animal Cruelty",
            "Bribery",
            "Sexual Assault With An Object"]

tableau_df = crimeD_df.loc[~crimeD_df.offense_name.isin(excluded)]
```
